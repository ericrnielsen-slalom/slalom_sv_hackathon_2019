### 1. Data Discovery

#### 1.1 Load Data and print basic stats

```python
# Load
raw_df = pd.read_csv('data/raw_data.csv')
```

```python
# Take a look at what the data looks like
pd.options.display.max_columns = None
display(raw_df)
```

```python
# Count number of rows and number of columns in the csv file
raw_df_row_count, raw_df_feature_count = raw_df.shape

# Print
print('Rows: %d, Columns: %d' % (raw_df_row_count, raw_df_feature_count))
```

#### 1.2 Form true test and dev sets

```python
# Choose train, dev, and test set percentages
train_split = 0.7
dev_split = 0.2
test_split = 1 - train_split

# Use choices above to determine size of sets
train_size = int(raw_df_row_count * train_split)
dev_size = int(raw_df_row_count * dev_split)
test_size = raw_df_row_count - train_size - dev_size

# Print
print('Training set size: %d' % train_size)
print('Dev set size: %d' % dev_size)
print('Test set size: %d' % test_size)
```

```python
# Randomly choose train, dev, and test sets based on percentage splits
train_set = raw_df.sample(n=train_size)
temp_df = raw_df.loc[~raw_df.index.isin(train_set.index)]
dev_set = temp_df.sample(n=dev_size)
test_set = temp_df.loc[~temp_df.index.isin(dev_set.index)]

# Make sure sizes make sense
train_rows, train_cols = train_set.shape
dev_rows, dev_cols = dev_set.shape
test_rows, test_cols = test_set.shape
assert(train_rows == train_size)
assert(dev_rows == dev_size)
assert(test_rows == test_size)
```

```python
# Save new train, dev, and test sets to csv files for easy access later
train_set.to_csv('data/train_data.csv', index=False)
dev_set.to_csv('data/dev_data.csv', index=False)
test_set.to_csv('data/test_data.csv', index=False)
```
